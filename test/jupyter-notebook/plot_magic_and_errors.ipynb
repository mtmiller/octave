```octave
%plot --format png
randn ("state", 1);
hist (randn (10000, 1), 30);
xlabel ("Value");
ylabel ("Count");
title ("Histogram of 10,000 normally distributed random numbers");
```

```octave
%plot -f svg
h = bar (rand (5, 10));
set (h(1), "basevalue", 0.5);
```

```octave
%plot --format jpg 
x = -10:0.1:10;
plot (x, sin (x));
title ("sin(x) for x = -10:0.1:10");
xlabel ("x");
ylabel ("sin (x)");
text (pi, 0.7, "arbitrary text");
legend ("sin (x)");
```
